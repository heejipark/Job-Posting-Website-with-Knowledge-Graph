```python
import pandas as pd
import os
import requests
import time
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

company = pd.read_csv('./data/companies.csv')
company.head(3)
```

```python
keywords = []
for c in company['company']:
    c = c.replace(' ', '+')
    c = c.replace('&','')
    keywords.append(c + '+glassdoor+overview')
keywords
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("headless")
new = np.arange(0,550,40)
urls = {}
error = []

for i, keyword in enumerate(keywords):
    if i in new:
        driver = webdriver.Chrome('/Users/heejipark/Documents/Knowledge-Graph/DSCI558_Project/chromedriver', options=chrome_options)
    print(i, keyword)
    driver.get('https://www.google.com/search?q='+keyword)
    time.sleep(0.5)
    html = driver.page_source
    soup = bs(html)
    r = soup.select('div.yuRUbf a')
    hrefs = [a['href'] for a in soup.find_all('a', href=True) if 'https://www.glassdoor.com/Overview/' in a['href']]
    if hrefs:
        hrefs = list(set(hrefs))
        urls[company.loc[i,'company']] = hrefs[0]
    else:
        error.append(company.loc[i,'company'])
urls
        
        
#     try:
#         links = soup.select('div.yuRUbf a', href=True)
#         url  = links['href']
#         print(url)
# #         if 'www.glassdoor.com' not in url:
# #             print("%d %s : error----"%(i,keyword))
# #             error.append(keyword)
# #             urls.append([company.loc[i,'company'],''])
# #         else:
# #             urls.append([company.loc[i,'company'], url])
#     except:
#         print("%d %s : error----"%(i,keyword))
#         error.append(keyword)
#         urls.append([company.loc[i,'company'],''])

#     if i+1 in new:
#         driver.quit()
```

```python
glassdoorURL = pd.DataFrame.from_dict(urls, orient='index')
glassdoorURL.reset_index(inplace=True)
glassdoorURL.columns = ['company','URL']
```

```python
glassdoorURL = company.merge(glassdoorURL, how='left', on='company')
glassdoorURL
```
