```python
import json
import pandas as pd

jsondata = './data/glassdoor_companyInfo.json'


with open(jsondata, 'r') as j:
     company = json.loads(j.read())
```

```python
len(company)
```

```python
company2 = pd.DataFrame.from_dict(company)
company2.to_csv('./data/glassdoor_companyInfo.csv', index=False)
```

```python
company = pd.DataFrame.from_dict(company).T.reset_index()
company = company.rename(columns = {'index':'company'})
company.to_csv('./data/glassdoor_companyInfo_T.csv', index=False)
```
